  <h2>Mostrar mensaje en pantalla:</h2>

  <p>Python nos permite mostrar mensajes en pantalla mediante la función: <code>print()</code> en la cual podemos escribir un simple texto entre comillas dentro de los paréntesis, pero en nuestro caso, habrá mensajes más interesantes que requieren <b>variables</b> o <b>constantes</b> como el número <b>PI</b> por ejemplo. Para esto, existe una convención al declarar variables: aunque para el lenguaje todas son variables, pero entre humanos, al leer un código ajeno, por "educación" las variables son escritas en minúsculas, mientras que las CONSTANTES son escritas en mayúsculas. En este lenguaje, para los que conocen <b>C</b>, no hace falta declarar el tipo de dato de la variable/constante, el intérprete, lo clasifica automáticamente, lo que no quiere decir que no existan. Además, estas no deben ser obligatoriamente declaradas al inicio: podemos ir inventándolas a medida que las requerimos. Sin embargo, la ejecución del código es secuencial (de arriba hacia abajo), y no podemos pretender invocar algo que aún no existe.
  Usaremos <b>f-strings</b> para un print más legible al indicar las variables o constantes, es muy simple, sólo requiere agregar la letra <b>f</b> antes de las comillas, esto nos permite mezclar, por así decirlo, el texto con las variables/CONSTANTES al encerrar estas entre llaves a medida que nuestro texto lo disponga. Ej: <code>print(f"La pendiente es: {m}")</code>
  Entonces, podemos comenzar declarando algunas variables:
<pre>

<h1>Entrada de usuario y bucles:</h1>
<p>
Como dijimos que declarábamos variables, pero en realidad indicamos un valor fijo, ahora pediremos al usuario el ingreso de datos. Esto lo realizamos con la función <code>input()</code> que le pedirá al usuario que ingrese lo pedido mediante el <b>string</b> el cual es una cadena de caracteres, encerrada entre comillas. Pero a diferencia de <b>print()</b> lo guardaremos inmediatamente en una variable (A partir de ahora no haré diferencia entre variable y CONSTANTE por cuestiones de lectura). <code>input()</code> nos devuelve siempre un <b>string</b>, sea lo que sea que ingrese el usuario. Nosotros debemos hacer la operación de conversión <b>(cast)</b> al tipo de datos que realmente requerimos. En nuestro caso usaremos <code>float()</code>(<i>decimal</i>). De esta manera, podemos presentar un entero como un decimal (1.0) y a la vez, manejar decimales.
</p><br>

<p>
Por otro lado, iniciaremos con la teoría de bucles, los cuales se utilizan para realizar una tarea repetidas veces. Pueden ser:
</p>

<ul>
<li>Si se cumple una condición: <code>if condición:</code>. Evalúa una condición que puede ser <b>booleana:</b> (<code>True</code> o <code>False</code>), si una variable es menor <code><</code> o mayor <code>></code> a otra, si son iguales <code>==</code> o distintas <code>!=</code>. Existen más condiciones a evaluar: ¡es cuestión de curiosidad!.</li>
</ul>
<br>

<ul>
<li>Mientras se dé la condición: <code>while condición:</code>. A veces también se usa la lógica inversa, mientras no se dé la condición: <code>while !condición:</code>.
</li>
</ul>
<br>

<ul>
<li>Una determinada cantidad de veces: <code>for i in range(4):</code>. La variable <code>i</code> (de iterator), permitirá almacenar el valor devuelto por la función <code>range()</code>, que nos devuelve los valores de 0 a 3. Por lo tanto, nos permite realizar la tarea <b>4 veces</b>, tras lo cual el bucle habrá terminado. Observá que ingresamos un 4, pero nos devuelve siempre 4 posiciones (0, 1, 2,  3)
</li>
</ul>
<br>

<ul>
<li>
El bucle <code>for iterador in algo_iterable:</code>, no solo usa <b>range()</b>. También puede iterar en otros tipos de datos como secuencias:
<ul>
<li><b>Listas:</b> <code>lista=["mate", "yerba", "azucar", 1, 3.4, 8, 8]</code></li>
<li><b>Tuplas:</b> <code>tupla=(12,"facturas")</code>.Estas, a diferencia de las listas, no se pueden modificar una vez declaradas.</li>
<li><b>Strings:</b> <code>texto="Buen día"</code></li>
<li><b>Rangos</b> generados por <code>range()</code> que solo acepta enteros, ya descrito</li>
</ul>
</li>
</ul><br>

<ul>
<li><b>Las colecciones, que existen 2 tipos:</b></li>
<li><b>Diccionarios:</b> Este es simple, en un dicionario que vas a buscar? Una palabra y su definición!.  Ejemplo: <code>funciones = {"Lineal": "y=mx+b", "Cuadratica":"ax<sup>2</sup>+bx+c","Polinomica":"ax<sup>impar</sup>+bx<sup>par</sup>+cx<sup>impar</sup>+...+d"}</code></li>
<li><b>Sets:</b> Conjuntos de elementos desordenados, que no admiten elementos duplicados <code>conjunto = {1, "uno", "phi", 99}</code></li>
</ul><br>

<p>Al realizar tareas repetitivas, normalmente las convertimos en funciones. En Python existen 2 tipos:
<ul>
<li><b>Anónimas</b> o <b>lambda</b>: que no poseen nombre y se usan para tareas simples. Ejemplo: <code>f = lambda x,m,b: m * x + b</code></li>
<li>Y las que si tienen nombre y usan la sintaxis: <code>def nombre(parámetros):</code>. Nuestra función puede o no requerir parámetros, así se llaman a las variables requeridas por una función, también llamadas argumentos y las funciones pueden devolver <code>return</code> algo o no devolver nada, dependiendo de lo que realicen.</li>
</ul>
</p>

<p>Ahora vamos a armar un segundo programa en el que apliquemos todo lo mencionado. Indicaremos mediante un mensaje al usuario que ingrese la pendiente (que verificaremos con un bucle <b>while</b>) y la ordenada al origen de una función lineal. Guardaremos la ecuación en variables <code>m</code> y <code>b</code>. También le pediremos un valor para evaluarla, el cual guardaremos en <code>x</code>. Además, mostraremos en pantalla su función y el resultado de la evaluación pedida. Como extra, le daremos los valores de los <code>x</code> previos, para utilizar <code>range()</code> con un bucle <code>for</code>. Pero como <b>range()</b> solo trabaja con números enteros (<b>int</b>), debemos castear la <b>x</b>:<code>int(x)</code>. Mediante el uso del bucle <code>if</code>, nos aseguraremos de que si el usuario ingresa una pendiente distinta de cero, nos saldremos del <b>while</b>. Si continúa ingresando cero, le pediremos nuevamente el valor de la pendiente.</p>

<p>El bucle <b>if</b> tiene una forma reducida llamada operador ternario. En la que usamos una sola línea de código para evaluar la condición con la siguiente sintaxis: <code>(valor si es True) if condición else (valor si es False)</code>.De esta manera evitamos escribir:

<pre>
if condición:
   hace algo
else:
   no hace nada
</pre>

No necesariamente debe o no hacer algo, es solo un ejemplo. El <code>else</code> se utiliza cuando queremos que, si no se cumple la condición del <code>if</code>, suceda otra tarea o indique un mensaje.

El bucle <b>if</b> se puede anidar. Te podés encontrar con la siguiente sintaxis:
<ul>
<li><code> if ...else</code></li>
<li><code> if ...if ...else</code></li>
<li><code> if ...else if ...else</code></li>
<li>O su forma abreviada: <code> if ...elif ...else</code></li>
<b><i>¡Ejercitar la curiosidad!.</i></b>
</ul>
</p>

<pre>
print("A continuación ingrese lo solicitado para evaluar una función lineal")

while True:
    m = float(input("\n¿Cuál es la pendiente de su función?: "))
    if m != 0:
        break
    print("\nPor favor, la pendiente de una función lineal no puede ser cero!\n")

b = float(input("\n¿Cuál es la ordenada al origen?: "))
x = float(input("\nIndique que valor desea evaluar en la función: "))

# Definimos la función
def lineal(m, x, b):
    return m * x + b

# Llamamos a la función y la guardamos en una variable
y = lineal(m, x, b)

# Mostramos la ecuación, usamos \t para tabular el texto.
print(f"\t\nLa función ingresada es: f(x) = {m}x + {b}\n")
print(f"\t Si evaluamos la función en f({x})={m}{x}+{b}\n")
print(f"\t Su resultado es: f({x})={y}\n")
print("")

# Le mostramos al usuario los resultados de evaluar los valores anteriores al que indicó para evaluar.
print("\n Estos son los valores de la función para los x anteriores a tu elección, desde 0 a x", (int(x) -1) if x > 0 else (int(x) + 1))

"""
En la línea anterior no usamos f-string,
presentamos el texto y separado por la coma el valor que nos devuelve
el operador ternario,
si x es positivo <code>x - 1</code> o negativo <code>x + 1</code>.
Además, vemos esta nueva sintaxis, comillas triples, para comentarios multílinea.
"""

if x > 0:
    for i in range(int(x)):
        print(f"f({i}) = {lineal(m, i, b)}")
if x < 0:
    for i in range(int(x)+1, 1):
        print(f"f({i}) = {lineal(m, i, b)}")
</pre>
